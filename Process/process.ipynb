```python
import pandas as pd
```

```python
data = pd.read_excel('月-品类.xlsx')
print(data)
```

```python
# 假设数据框的列名为 '品类名称'
# 选择出品类名称相同的数据
same_category_data = data[data.duplicated(subset=['品类名称'], keep=False)]

# 如果你只想保留重复品类名称的第一行数据，可以将 keep 参数设置为 'first'
# same_category_data = df[df.duplicated(subset=['品类名称'], keep='first')]

# 如果你想删除重复品类名称的行，可以使用 drop_duplicates 方法
# df.drop_duplicates(subset=['品类名称'], keep=False, inplace=True)

# 打印筛选出的数据
print(same_category_data)
```

```python

# 使用 groupby 根据品类名称进行分组
grouped_data = data.groupby('品类名称')

# 创建一个字典，用于存储分组后的数据框
category_data_dict = {}

# 遍历分组后的数据，将每个品类名称的数据存储在字典中
for category, group in grouped_data:
    category_data_dict[category] = group

# 打印示例品类的数据框
print(category_data_dict['水生根茎类'])
```

```python
import pandas as pd

# 使用 groupby 根据品类名称进行分组
grouped_data = data.groupby('品类名称')

# 创建一个 ExcelWriter 对象来写入 Excel 文件
with pd.ExcelWriter('category_data.xlsx') as writer:
    # 遍历分组后的数据，将每个品类名称的数据存储在单独的工作表中
    for category, group in grouped_data:
        # 将每个品类的数据写入到单独的工作表中
        group.to_excel(writer, sheet_name=category, index=False)

# 文件保存完成后，会自动关闭 ExcelWriter 对象

```

```python
import pandas as pd

# 使用 groupby 根据品类名称进行分组
grouped_data = data.groupby('品类名称')

# 遍历分组后的数据，将每个品类名称的数据保存到不同的 Excel 文件中
for category, group in grouped_data:
    # 创建一个 ExcelWriter 对象来写入单独的 Excel 文件
    excel_filename = f"{category}.xlsx"
    with pd.ExcelWriter(excel_filename) as writer:
        # 将品类的数据写入到 Excel 文件中
        group.to_excel(writer, sheet_name='Sheet1', index=False)
```
