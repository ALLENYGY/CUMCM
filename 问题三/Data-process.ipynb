```python
# 将销售日期列转换为日期时间类型
df['销售日期'] = pd.to_datetime(df['销售日期'])

# 筛选出日期在2023-06-24到2023-06-30之间的数据
start_date = '2023-06-24'
end_date = '2023-06-30'
filtered_df = df[(df['销售日期'] >= start_date) & (df['销售日期'] <= end_date)]
unique_product_names = filtered_df['单品名称'].unique()
# 打印不同单品名称的数量
print("不同的单品名称数量:", len(unique_product_names))
# 打印不同的单品名称
unique_product_names
```

```python
desired_products = unique_product_names

start_date = '2023-04-01'
end_date = '2023-06-30'
filtered_data = df[(df['销售日期'] >= start_date) & (df['销售日期'] <= end_date)]
# 使用isin()函数筛选单品名称在desired_products列表中的数据
filtered_data = filtered_data[filtered_data['单品名称'].isin(desired_products)]
filtered_data
```

```python

# 你需要读取该文件并创建一个DataFrame
df = pd.read_excel('Q3.xlsx')

# 将日期列转换为日期时间对象
df['销售日期'] = pd.to_datetime(df['销售日期'])

# 将销售日期作为索引
df.set_index('销售日期', inplace=True)

# 使用groupby按单品名称分组，并计算每日总销售量
daily_sales = df.groupby('单品名称')['销量(千克)'].resample('D').sum().reset_index()

# 打印结果
daily_sales
```

```python
df = pd.read_excel('Q3.xlsx')
df.drop(columns=['Unnamed: 0'], inplace=True)
df.drop(columns=['销量(千克)'], inplace=True)
daily_sales= pd.merge(daily_sales, df, left_on=['单品名称','销售日期'], right_on=['单品名称','销售日期'], how='left')
daily_sales.to_excel('daily_sales.xlsx')
daily_sales
```

```python

filtered_df = daily_sales.groupby(['销售日期', '单品名称']).first().reset_index()
```

```python
filtered_df.sort_values(by=['单品名称','销售日期'], inplace=True)
filtered_df
```

```python
filtered_df=filtered_df[filtered_df['销量(千克)']>0]

filtered_df['销售利润(元/千克)'] = (filtered_df['销售单价(元/千克)']-filtered_df['批发价格(元/千克)'])*(1-filtered_df['损耗率(%)']/100)
filtered_df['单日销售利润'] = filtered_df['销售利润(元/千克)']*filtered_df['销量(千克)']
#  销售量 * (销售单价 - 批发价格) * (1 - 损耗率)
filtered_df.to_excel('Q3-1.xlsx')
filtered_df
```
