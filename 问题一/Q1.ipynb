```python
import pandas as pd

df=pd.read_excel('/Users/allenygy/Desktop/WorkSpace/VScode-Workspace/CODE/Python/国赛/Data/Table.xlsx')
df
```

```python
# 提取年份和月份
df['年'] = df['销售日期'].dt.year
df['月'] = df['销售日期'].dt.month
df
```

```python
category = df.groupby(['年','月','分类名称'])['销量(千克)'].sum().reset_index()
category['日期'] = pd.to_datetime(category['年'].astype(str) + '-' + category['月'].astype(str) + '-01')
category
```

# 分类名称

```python
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

df = category

# 按品类和日期分组并计算每个月的销售总量
monthly_sales = df.groupby(['分类名称', pd.Grouper(key='日期', freq='M')])['销量(千克)'].sum().reset_index()

# 创建一个包含六个子图的图形
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 15))
fig.suptitle('每月销售总量折线图按品类')

# 循环绘制六个品类的折线图
for i, (category, data) in enumerate(monthly_sales.groupby('分类名称')):
    row, col = divmod(i, 3)
    ax = axes[row, col]
    ax.plot(data['日期'], data['销量(千克)'], marker='o', linestyle='-')
    ax.set_title(category)
    ax.set_xlabel('日期')
    ax.set_ylabel('销量(千克)')
    ax.grid(True)

# 调整子图布局
plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
# 按品类和日期分组并计算每个月的销售总量
monthly_sales = df.groupby(['分类名称', pd.Grouper(key='日期', freq='M')])['销量(千克)'].sum().reset_index()

# 创建一个图形
plt.figure(figsize=(15, 8))
plt.title('每月销售总量折线图按品类')

# 使用不同颜色循环绘制六个品类的折线图
colors = ['b', 'g', 'r', 'c', 'm', 'y']  # 可以根据需要选择颜色
for i, (category, data) in enumerate(monthly_sales.groupby('分类名称')):
    plt.plot(data['日期'], data['销量(千克)'], marker='o', linestyle='-', label=category, color=colors[i])

plt.xlabel('日期')
plt.ylabel('销量(千克)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 使用apply函数将'年'、'月'和'单品名称'列拼接成一个新的列
df['年-月-品类']  = df.apply(lambda row: f"{row['年']}年{row['月']}月{row['分类名称']}", axis=1)
df.to_excel('category.xlsx', index=False)
```
