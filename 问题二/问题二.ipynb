```python
import pandas as pd
df = pd.read_excel('/Users/allenygy/Desktop/WorkSpace/VScode-Workspace/CODE/Python/国赛/问题一/Result-process.xlsx')
df
```

```python
data=df


data['销量占比'] = data.groupby(['销售日期', '分类名称'])['销量(千克)'].transform(lambda x: x / x.sum())

print(data)
data.to_excel('data.xlsx')
```

```python
# 计算每个品类每天的加权售价
category_daily_weighted_price = (data['销售单价(元/千克)'] * data['销量占比']).groupby([data['销售日期'], data['分类名称']]).sum().reset_index()
category_daily_weighted_price.fillna(0, inplace=True)
# 打印结果
print(category_daily_weighted_price)

category_daily_weighted_price.to_excel('category_daily_weighted_price.xlsx')
```
